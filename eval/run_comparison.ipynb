# Estudo comparativo Cíclope × ARGUS — Fase 5

Rode **no servidor** (ambiente com ML) para incluir o **ARGUS** (precisa do detector E1).
No Mac (LITE) o ARGUS é **pulado** automaticamente e só o Cíclope roda.

Use o **kernel do venv do backend** (onde `app` e as deps de ML estão instaladas).
Resultados (com **cache**) ficam em `eval/results/`.

```python
import os, sys
# sobe até a raiz do projeto (onde existem eval/ e backend/)
for _ in range(6):
    if os.path.isdir('eval') and os.path.isdir('backend'):
        break
    os.chdir('..')
sys.path.insert(0, os.path.abspath('eval'))
import harness, metrics
from app.argus import detect
print('cwd =', os.getcwd())
print('ARGUS disponível (detector)?', detect.available())
```

```python
# Configuração do estudo
N = 3
SYSTEMS = ('ciclope', 'argus')
IMAGES = harness.default_images()   # 2 Figuras do enunciado; troque pela lista do gold set (5.3)
[p.name for p in IMAGES]
```

```python
# Execução (sequencial, com cache). force=True ignora o cache e re-roda.
runs = harness.run_matrix(IMAGES, systems=SYSTEMS, n=N, force=False)
measured = harness.measure_runs(runs)
summary = metrics.summarize(measured)
print(metrics.format_table(summary))
```

```python
# Salva a tabela (média±desvio) em eval/results/summary.md
out = harness.RESULTS_DIR
out.mkdir(parents=True, exist_ok=True)
(out / 'summary.md').write_text(metrics.format_table(summary) + '\n', encoding='utf-8')
print('salvo em', out)
```

**Notas.** A variância entre as N execuções é a coluna `±` de cada métrica (a não-determinância do VLM, agora medida). Esta é a régua única (`app.compare.measure`) — os mesmos números do painel "Comparar". Os lotes 5.2 (juiz) e 5.3 (gold set) entram depois.
